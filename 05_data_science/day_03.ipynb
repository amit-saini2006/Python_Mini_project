```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

### Load CSV Data
To do it we can simply use pandas library

```python
data = pd.read_csv("experience_salary.csv")
```

### Fit model
Enter the data in model

```python
x = data[["YearsExperience"]]
y = data[["Salary"]]

model = LinearRegression()
model.fit(x, y)

data['PredictedSalary'] = model.predict(x)
print("Model Coefficeint (slope)", round(model.coef_[0][0], 2))
print("Model Intercept (base salary)", round(model.intercept_[0], 2))
```

### Plotting Graph

```python
plt.scatter(x, y, color='blue', label="Actual Data")
plt.plot(x, data['PredictedSalary'], color='red', label='Regression line')
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.title("Salary vs Experience")
plt.legend()
plt.grid(True)
plt.tight_layout()
```
